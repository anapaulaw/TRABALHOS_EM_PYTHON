# Quando o crédito deve ser aprovado?


O objetivo  é criar um programa simples que possa prever se um crédito deve ser aprovado ou não, com base em dados de clientes anteriores. O programa deve ajudar o banco a minimizar o risco com futuros clientes.

O modelo utilizado irá prever se um cliente deve ter seu empréstimo aprovado ou não, com base em seu histórico, utilizando regressão logística para classificação.


IMPORTANDO BIBLIOTECAS

```python
import numpy as np
import pandas as pd 
from matplotlib import pyplot as plt
import seaborn as sns
import os
import sklearn
from sklearn.model_selection import train_test_split
```

CARREGANDO BANCO DE DADOS

```python
data = pd.read_csv('/content/german_credit (1).csv')
data.head()
```

ESTATÍSTICAS DESCRITIVAS DOS DADOS

```python
data.describe()
```

CORRELAÇÃO DAS VARIÁVEIS

```python
data.corr()
```

Em seguida selecionamos as variáveis  com maior correlação que foram o Saldo da conta, Duração do crédito (mês),Status de pagamento do crédito anterior e separar em um dataframe separado da  variável dependente Creditabilidade.

```python
x = data[['Account Balance', 'Duration of Credit (month)', 'Payment Status of Previous Credit']]

y = data['Creditability']
```

DIVISÃO DOS DADOS EM TESTE E TREINO.
80% TESTE 
20% TREINO

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.80, random_state=6)
```

AJUSTANDO OS DADOS DE TREINAMENTO E APLICANDO MODELO ESTATISTICA PARA GERAR AS PROBABILIDADES.


```python
import statsmodels.api as sm
```

```python
model = sm.Logit(y_train, x_train).fit()
```

```python
model.summary()
```

VALORES PREDITO PELO MODELO



```python
pred_y = model.predict(x_test) 
pred_y.head()
```

FUNÇÃO PARA TRANSFORMAR EM PROBABILIDADE

```python
def binary_classify(x): 
    x = round(x, 2)
    if x >= 0.50:
        return 1
    return 0

pred = list(map(binary_classify, pred_y)) 
```

ACURÁCIA DO MODELO

```python
sklearn.metrics.accuracy_score(y_test, pred)
```

O MODELO ACERTOU 73,75% DAS PREVISÕES.

Solução proposta
Para classificar os futuros clientes como aptos ou não a concessão de crédito conforme entrada de dados sobre entradas o saldo da conta, a duração do crédito e a situação de pagamento do crédito atual. O programa então aplicará o modelo anterior e classificará o cliente, informando ao banco se o crédito deve ser aprovado ou não.

FUNÇÃO PREDITIVA BASEADA EM MODELO LOGISTICO - RETORNA PROBABILIDADE

```python
def log_func(x):
    balance, credit, pay_status = x
    result = (0.6364 * balance) + (-0.05 * credit) + (0.2374 * pay_status)
    return result

def binary_classify(x): 
    x = round(x, 2)
    if x >= 0.50:
        return 1
    return 0

def predict(x): 
    return binary_classify(log_func(x)) 

def print_result(x): 
    if predict(x) == 1:
        return "digno de crédito"
    return "Não digno de crédito"
```

VERIFICANDO DADOS DE UM CLIENTE ESPECIFICO


Dados do cliente:

```python
input_test = (2, 36, 3) # client data
print('Result: ' + print_result(input_test))
```
